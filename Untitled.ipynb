```python
from skimage import io
import matplotlib.image as img
from sklearn.cluster import KMeans
import numpy as np
from matplotlib import pyplot as plt

image = io.imread("IMG_1075.JPG")
X = image.reshape(-1,3)  #-133

# 
mod = KMeans(n_clusters = 5)  #
labels = mod.fit_predict(X)  #0-31
colors = mod.cluster_centers_  #32

# 
new_img = colors[labels]
new_img = new_img.reshape(image.shape)  #
new_img = new_img.astype(np.uint8)
plt.imshow(new_img)
img.imsave('picture_new.jpg',new_img)
plt.show()

```

```python
import colorsys
from tqdm import tqdm
from PIL import Image

# 

filename = 'IMG_1075.JPG'

# 

target_hue = 200

#  RGB 

image = Image.open(filename).convert('RGBA')

#  RGB 

image.load()

r, g, b, a = image.split()

result_r, result_g, result_b, result_a = [], [], [], []

# 

for pixel_r, pixel_g, pixel_b, pixel_a in tqdm(zip(r.getdata(), g.getdata(), b.getdata(), a.getdata())):
    #  HSV 
    h, s, v = colorsys.rgb_to_hsv(pixel_r / 255., pixel_b / 255., pixel_g / 255.)
    #  RGB 
    rgb = colorsys.hsv_to_rgb(target_hue/360, s, v)
    pixel_r, pixel_g, pixel_b = [int(x * 255.) for x in rgb]
    # 
    result_r.append(pixel_r)
    result_g.append(pixel_g)
    result_b.append(pixel_b)
    result_a.append(pixel_a)

r.putdata(result_r)
g.putdata(result_g)
b.putdata(result_b)
a.putdata(result_a)

# 

image = Image.merge('RGBA', (r, g, b, a))

# 

image.save('output.png')
```

```python
import os
import colorsys
import Image

def hueChange(img, hue):
    # It's better to raise an exception than silently return None if img is not
    # an Image.
    img.load()
    r, g, b = img.split()
    r_data = []
    g_data = []
    b_data = []

    for rd, gr, bl in zip(r.getdata(), g.getdata(), b.getdata()):
        h, s, v = colorsys.rgb_to_hsv(rd / 255., bl / 255., gr / 255.) 
        rgb = colorsys.hsv_to_rgb(hue/360., s, v)
        rd, gr, bl = [int(x*255.) for x in rgb]
        r_data.append(rd)
        g_data.append(gr)
        b_data.append(bl)

    r.putdata(r_data)
    g.putdata(g_data)
    b.putdata(b_data)
    return Image.merge('RGB',(r,g,b))

filename = 'IMG_1075.JPG'
basename, ext = os.path.splitext(filename)
img = Image.open(filename).convert('RGB')
for hue in range(1, 360, 30):
    img2 = hueChange(img, hue)
    out_name = '{}_hue{:03d}.jpg'.format(basename, hue)
    img2.save(out_name)
```
